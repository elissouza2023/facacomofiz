```python
Y = peso_laranja
X = diametro_laranja
n = np.size(X)

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
b = np.mean(Y) - a*np.mean(X)
```

Para ajudar no entendimento vamos criar dois gráficos com as retas, usando os dados das toranjas e laranjas.

🧠 Dicas para interpretar o gráfico:
A dispersão dos pontos mostra a variação do peso com o diâmetro para cada fruta.

A reta mostra a tendência geral: se ela sobe, quer dizer que quanto maior o diâmetro, maior o peso.

Compare a inclinação das retas:

Se a reta da toranja for mais inclinada, significa que o peso aumenta mais rapidamente com o diâmetro do que na laranja.

```python
import numpy as np
import matplotlib.pyplot as plt

# Supondo que 'dado' já esteja carregado e seja um array NumPy
diametro_laranja = dado[:5000, 0]
peso_laranja = dado[:5000, 1]

diametro_toranja = dado[5000:, 0]
peso_toranja = dado[5000:, 1]

# Função para calcular os coeficientes a e b
def calcular_coeficientes(X, Y):
    n = np.size(X)
    a = (n * np.sum(X * Y) - np.sum(X) * np.sum(Y)) / (n * np.sum(X**2) - np.sum(X)**2)
    b = np.mean(Y) - a * np.mean(X)
    return a, b

# Calcular para laranja
a_laranja, b_laranja = calcular_coeficientes(diametro_laranja, peso_laranja)

# Calcular para toranja
a_toranja, b_toranja = calcular_coeficientes(diametro_toranja, peso_toranja)

# Gerar os valores das retas
x_vals_laranja = np.linspace(min(diametro_laranja), max(diametro_laranja), 100)
reta_laranja = a_laranja * x_vals_laranja + b_laranja

x_vals_toranja = np.linspace(min(diametro_toranja), max(diametro_toranja), 100)
reta_toranja = a_toranja * x_vals_toranja + b_toranja

# Criar o gráfico
plt.figure(figsize=(10, 6))

# Dispersão dos dados
plt.scatter(diametro_laranja, peso_laranja, color='orange', alpha=0.5, label='Laranjas 🍊')
plt.scatter(diametro_toranja, peso_toranja, color='green', alpha=0.5, label='Toranjas 🍈')

# Retas de regressão
plt.plot(x_vals_laranja, reta_laranja, color='darkorange', linewidth=2, label='Reta Laranja')
plt.plot(x_vals_toranja, reta_toranja, color='darkgreen', linewidth=2, label='Reta Toranja')

# Legenda e labels
plt.title('Diâmetro vs Peso - Laranjas e Toranjas')
plt.xlabel('Diâmetro (cm)')
plt.ylabel('Peso (g)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Para ajudar a entender criamos os graficos separados

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados
diametro_laranja = dado[:5000, 0]
peso_laranja = dado[:5000, 1]

diametro_toranja = dado[5000:, 0]
peso_toranja = dado[5000:, 1]

# Função para calcular os coeficientes a e b
def calcular_coeficientes(X, Y):
    n = np.size(X)
    a = (n * np.sum(X * Y) - np.sum(X) * np.sum(Y)) / (n * np.sum(X**2) - np.sum(X)**2)
    b = np.mean(Y) - a * np.mean(X)
    return a, b

# Calcular para laranja
a_laranja, b_laranja = calcular_coeficientes(diametro_laranja, peso_laranja)
x_vals_laranja = np.linspace(min(diametro_laranja), max(diametro_laranja), 100)
reta_laranja = a_laranja * x_vals_laranja + b_laranja

# Calcular para toranja
a_toranja, b_toranja = calcular_coeficientes(diametro_toranja, peso_toranja)
x_vals_toranja = np.linspace(min(diametro_toranja), max(diametro_toranja), 100)
reta_toranja = a_toranja * x_vals_toranja + b_toranja

# Criar dois subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico 1: Laranjas 🍊
axs[0].scatter(diametro_laranja, peso_laranja, color='orange', alpha=0.5, label='Laranjas')
axs[0].plot(x_vals_laranja, reta_laranja, color='darkorange', linewidth=2, label='Reta Laranja')
axs[0].set_title('Laranjas 🍊')
axs[0].set_xlabel('Diâmetro (cm)')
axs[0].set_ylabel('Peso (g)')
axs[0].legend()
axs[0].grid(True)

# Gráfico 2: Toranjas 🍈
axs[1].scatter(diametro_toranja, peso_toranja, color='green', alpha=0.5, label='Toranjas')
axs[1].plot(x_vals_toranja, reta_toranja, color='darkgreen', linewidth=2, label='Reta Toranja')
axs[1].set_title('Toranjas 🍈')
axs[1].set_xlabel('Diâmetro (cm)')
axs[1].set_ylabel('Peso (g)')
axs[1].legend()
axs[1].grid(True)

plt.suptitle('Análise Separada: Diâmetro vs Peso', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
